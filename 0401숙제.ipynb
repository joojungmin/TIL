# 2. 369 게임

```python
# for문으로 숫자자리 중에 3 or 6 or 9 가 있으면 그 개수만큼 'clap' 출력
lastNumber = int(input("마지막 숫자를 입력하세요: "))

for i in range(1,lastNumber):
    clap_cnt = 0
    for j in str(i):
        if j in '369':
            clap_cnt +=1
    if clap_cnt == 0:
        print(i,": ", i)
    else:
        print(i, ": ", "clap!"*clap_cnt)
```
